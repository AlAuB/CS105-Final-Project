```python
# import packages
import numpy as np
import pandas as pd
```

```python
#Read .csv file
#Dataset of EVs that are currently registered through The State of Washington.
data = pd.read_csv('Electric_Vehicle_Population_Data.csv')
data.head(10) #Show first 10 lines
```

## Dataset Info

```python
data.info() #Get the property of the dataset
```

## Drop Non-Washington Data

```python
data_filtered = data[data['State'] == 'WA']
data_filtered
```

```python
data_filtered.shape
```

```python
data.shape[0] - data_filtered.shape[0] #Num of rows that be filtered out
```

## Top 10 Cities in WA with EVs

```python
top_city = data_filtered.groupby('City').size().reset_index(name='Count')
top_city.sort_values(by='Count', inplace=True, ascending=False)
top_city[:10]
```

## How Many EVs for Each Brand

```python
make_count = data['Make'].value_counts() #Show how many EVs each company made.
make_count
```

```python
pie_plot = (100 * make_count[:10] / len(make_count)).plot(kind = 'pie', autopct='%1.1f%%', title='Evs Per Brand') #Only display the top 10 brands
```

## EVs per Model Year

```python
ev_per_model_year = data_filtered.groupby('Model Year').size().reset_index(name='Count')
ev_per_model_year.plot.bar(x='Model Year', y='Count')
```

```python
plot = (100 * data_filtered['Model Year'].value_counts()[:10] / len(data_filtered['Model Year'])).plot(kind='pie', autopct='%1.1f%%', title='EVs per Model Year (%)')
```

## EVs per Type

```python
type_count = data_filtered['Electric Vehicle Type'].value_counts()
type_count.plot.barh()
```

## Top 10 Popular Model

```python
popular_model = data_filtered.dropna()
popular_model = popular_model.groupby(['Make', 'Model', 'Electric Vehicle Type']).size().reset_index(name='Count')
popular_model.sort_values(by='Count', inplace=True, ascending=False)
popular_model[:10]
```

## Top 10 Electric Range EVs

```python
electric_range = data_filtered.sort_values(by="Electric Range", ascending=False)
electric_range = electric_range.drop_duplicates(subset='Electric Range', keep='first')
electric_range = electric_range.drop_duplicates(subset=['Model Year', 'Make', 'Model'], keep='first')
electric_range = electric_range[['Model Year', 'Make', 'Model', 'Electric Range']]
electric_range[:10]
```
